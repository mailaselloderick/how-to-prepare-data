```python
import pandas as pd
```

```python
#1.Read the data from the .csv file into a DataFrame and display the first five rows
df_avocado=pd.read_csv('avocado.csv')
df_avocado.head()
```

```python
#2.Notice that the column names are inconsistent.
# Change the column names so they all use title case with no spaces.
df_avocado.columns=df_avocado.columns.str.title().str.replace(' ','')
```

```python
#3.Add a column with the percentage of the bags in the TotalBags column that are extra-large bags.
df_avocado['percent_of_Xlbags']=(df_avocado.XlargeBags / df_avocado.TotalBags) * 100
```

```python
#4.Add a column with the percentage of the bags in the TotalBags column that are large bags
df_avocado['percent_of_Lbags']=(df_avocado.LargeBags / df_avocado.TotalBags) * 100
```

```python
#5.Add a column with the percentage of the bags in the TotalBags column that are small bags.
df_avocado['percent_of_Sbags']=(df_avocado.SmallBags / df_avocado.TotalBags) * 100
```

```python
#6.Display the first five rows of data to view the new columns
df_avocado.head()
```

```python
#7.Assign the Region, Type, Year, and TotalBags columns to a new DataFrame.
df_avocado=df_avocado[['Region','Type','Year','TotalBags']]
df_avocado.head()
```

```python
#8.Group, sum, and unstack this data to produce a DataFrame that looks like this.
columns_grouped=df_avocado.groupby(['Region','Type','Year'])[['TotalBags']].sum()
column_unstack=columns_grouped.unstack('Type')
column_unstack
```
