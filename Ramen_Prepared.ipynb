```python
import pandas as pd
```

```python
#1.Read the data from the CSV file into a DataFrame and display the first five rows
df_ramen=pd.read_csv('ramen-ratings.csv')
df_ramen.head()
```

```python
#2.Add a column to the DataFrame that stores the average rating for each country.
# Then, display the first five rows and verify that the column was added and
# that its values appear to be correct.
df_ramen['average_rating']=df_ramen.groupby('Country')['Stars'].transform('mean')
df_ramen.head()
```

```python
#3.Group the data by the Country and Style columns and take the mean. 
# Then,unstack the Style column and display the data for the Stars column
columns_grouped=df_ramen.groupby(['Country','Style'])['Stars'].mean()
column_unstack=columns_grouped.unstack('Style')
column_unstack.head()
```

```python
#Write a function that takes a row as input and returns a shortened version of the values in the Variety column.
# The values should consist of the first two words of the variety name, followed by a space and an ellipsis (...). 
# Hint: Youâ€™ll need to split the values into a list of words, get the first two words in the list,
# add the ellipsis to the list, and then join the words in the list.
def shortened_version(row):
    
    words=row.Variety.split()
    word = ' '.join(words[:2]) + "..."
    return word
```

```python
#5.Apply the function you just wrote to the DataFrame,and store the results in a column named ShortVariety.
# Then, display the first five rows to be sure this worked
df_ramen['short_variety'] = df_ramen.apply(shortened_version, axis=1)
df_ramen.head()
```

```python
#6.Set an index on the Brand column and then display the first five rows of data
df_ramen.set_index('Brand',inplace=True)
df_ramen.head()
```
